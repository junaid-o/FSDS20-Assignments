### Question 01:
What does RGBA stand for?

**<span style='color:blue'>Answer</span>**

RGBA stands for Red, Green, Blue, and Alpha. It is a color model that is commonly used in computer graphics and image processing to represent and specify colors. The RGBA model extends the RGB model by adding an additional component, the Alpha channel, which represents the transparency or opacity of the color.

### Question 02:
From the Pillow module, how do you get the RGBA value of any images?

**<span style='color:blue'>Answer</span>**

To get the RGBA value of an image using the Pillow module in Python, you can use the `getpixel()` method of an `Image` object. The `getpixel()` method takes the coordinates `(x, y)` of the pixel you want to retrieve, and it returns a tuple containing the RGBA values of that pixel. You can access the individual color components by indexing the tuple.

```python
from PIL import Image


image = Image.open("example.png")

# Get the RGBA value of a specific pixel
pixel = image.getpixel((x, y))

# The pixel variable will contain a tuple of RGBA values
# Access individual color components like this:
red = pixel[0]
green = pixel[1]
blue = pixel[2]
alpha = pixel[3]
```

### Question 03:
What is a box tuple, and how does it work?

**<span style='color:blue'>Answer</span>**

In the context of the Pillow library in Python, a box tuple represents a rectangular region within an image. It is a tuple of four values: `(left, upper, right, lower)`. Each value represents a coordinate that defines the boundaries of the box.

- `left`: The x-coordinate of the left edge of the box.
- `upper`: The y-coordinate of the upper edge of the box.
- `right`: The x-coordinate of the right edge of the box (not inclusive).
- `lower`: The y-coordinate of the lower edge of the box (not inclusive).

The box tuple is used in various functions and methods in Pillow to specify regions of interest within an image, such as cropping, resizing, or applying operations to specific areas.

### Question 08:
Image objects do not have drawing methods. What kind of object does? How do you get this kind
of object?

**<span style='color:blue'>Answer</span>**

To get an ImageDraw object, you need to call the ImageDraw.Draw() function and pass the Image object as the argument. 

````python
from PIL import Image, ImageDraw


image = Image.open("pikachu.png")

# Get the ImageDraw object
draw = ImageDraw.Draw(image)

# Use the draw object to perform drawing operations
draw.line([(0, 0), (100, 100)], fill="red", width=2)
draw.rectangle([(50, 50), (150, 150)], outline="blue", width=2)
draw.text((200, 200), "Hello, World!", fill="green")


################################
image.save("modified.jpg")
````
